```python
import random
s=0
c=0
d=0
while(1):
    computer=random.choice([-1,0,1])
    youdict={-1:"SNAKE",0:"WATER",1:"GUN",10:"exit"}
    print(youdict)
    you=int(input("enter your choice"))
    if you==10:
        print("thank you for playing game")
        print(f"your score\t\tcomputer score\t\tdraws\n  \t{c}\t\t{s}\t\t\t{d}")
        break 
    if(you!=1 and you!=-1 and you!=0 and you!=10):
        print("invalid choice")
    print(f"you chose-{youdict[you]}\ncomputer chose-{youdict[computer]}")

    if computer==you:
        print("it's draw")
        d+=1

    elif computer!=you:
        if (computer==-1 and you==0) or (computer==0 and you==1) or (computer==1 and you==-1):
            print("you lost!")   
            s+=1
        elif (computer==0 and you==-1) or (computer==1 and you==0) or (computer==-1 and you==1): 
            print("you won!!!")
            c+=1
       
          
```
