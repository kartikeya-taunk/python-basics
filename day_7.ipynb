```python
# recursion --it is a function which calls it-self
def factorial(n):
    if n==1 or n==0:
        return 1
    return n*factorial(n-1)
n=int(input("enter no.:"))
print(f"the factorial is {factorial(n)}")    
    
```

```python
def natural(n):
    
    if n==0:
        return 0
    return n+natural(n-1)
n=int((input(" enter number:")))
print(f"the sum is {natural(n)}")  
```

```python
def rev(l,word):
    n=[]
    for i in l:
        if not(i==word):
            n.append(i.strip(word))
    return n
l=["kartikeya","rohan","an","modi"] 
print(rev(l,"an"))      


    
```
