```python
#how to access the element from the list
num=[10,20,30,40]
num[0]
num[2]
num[-1]
num[-2]
```

```python
#list slicing
num[0:3] #2 elements #0,1,2,(0 excluded)
num[-1:-3:-1]# num[start:stop:step], it takes 1 on step as default
num[:]
num.append([50,60])
num
```

```python
#list modify
list1=[10,20,30,40]
list1
list1[0]=100
list1
```

```python
#add elements
num=[1,2,3,4]
num.append(5)#add at the last of the list
num
```

```python
#extend
num.extend([2,3,4])
num
```

```python
#insert (it add the element at any place acc to the given index)
num.insert(6,2)#(index,value)
num
```

```python
#removing elements
num.remove(5)#it will remove the first occurance of given element
num
```

```python
#pop()
num.pop()#it will remove the element from the last of the list
num
num.pop(3)#it will remove the index value element
num
```
