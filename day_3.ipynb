```python
#tuple unpacking
t=(1,2,3)
a,b,c=t
print(a)
print(b)
print(c)
```

```python
#set operations
#union,intesection,difference,symmetic difference
a={1,2,3}
b={3,4,5}
c={4,3,6}
print(a.union(b,c))
```

```python
print(a|b|c)
```

```python
a.intersection(b,c)
```

```python
print(a&b)
```

```python
#difference (element of a but not b,also not common one)
print(a-b)
print(b-c)
print(b-a)
```

```python
#symmetic difference
print(a^b)
print(c^b)
print(a^c)
print(b^a)
```

```python
#set immutable no change
#frozenset()
fs=frozenset(a)
print(a)
print(fs)
#fs.update({3,4,9,11,22})# error because its frozen
print(fs)
print(a)
```
